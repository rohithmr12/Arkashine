```python
import numpy as np
from scipy.signal import savgol_filter

def msc(input_data, reference=None):
    if reference is None:
        reference = np.mean(input_data, axis=0)
    corrected_data = np.zeros_like(input_data)
    for i in range(input_data.shape[0]):
        fit = np.polyfit(reference, input_data[i, :], 1, full=True)
        corrected_data[i, :] = (input_data[i, :] - fit[0][1]) / fit[0][0]
    return corrected_data

def first_derivative(input_data, window_length=11, polyorder=2):
    return savgol_filter(input_data, window_length=window_length, polyorder=polyorder, deriv=1)

def second_derivative(input_data, window_length=11, polyorder=2):
    return savgol_filter(input_data, window_length=window_length, polyorder=polyorder, deriv=2)

def snv(input_data):
    return (input_data - np.mean(input_data, axis=1, keepdims=True)) / np.std(input_data, axis=1, ddof=1, keepdims=True)

df=df.to_numpy()
msc_data = msc(df)
fd_data = first_derivative(msc_data)
df = snv(fd_data)

```

```python

from sklearn.model_selection import cross_val_score, KFold
from sklearn.cross_decomposition import PLSRegression
```

```python
df1.fillna(df1.mean(), inplace=True)
```

```python
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python

def select_optimal_wavelengths(X, Y, n_components=10):
    pls = PLSRegression(n_components=n_components)
    pls.fit(X, Y)
    loadings = pls.x_loadings_
    selected_wavelengths = np.argsort(np.abs(loadings), axis=0)[-1,:] 
    return selected_wavelengths, loadings


selected_wavelengths_dict = {}

for nutrient_idx in range(df1.shape[1]):
    nutrient_name = df1.columns[nutrient_idx]
    Y_nutrient = df1.iloc[:, nutrient_idx].values.reshape(-1, 1)
    
    selected_wavelengths, loadings = select_optimal_wavelengths(df, Y_nutrient)
    selected_wavelength_names = [f"Wavelength_{i+1}" for i in selected_wavelengths]
    
    selected_wavelengths_dict[nutrient_name] = selected_wavelength_names
    print(f"Selected Wavelengths for {nutrient_name}: {selected_wavelength_names}")

    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    pls = PLSRegression(n_components=10)
    cv_scores = cross_val_score(pls, df, Y_nutrient, cv=kf, scoring='neg_mean_squared_error')
    
    print(f"Cross-Validation MSE Scores for {nutrient_name}:", -cv_scores)
    print(f"Average Cross-Validation MSE for {nutrient_name}:", -np.mean(cv_scores))
    print()

df_selected_wavelengths = pd.DataFrame.from_dict(selected_wavelengths_dict, orient='index')
print(df_selected_wavelengths)
```
